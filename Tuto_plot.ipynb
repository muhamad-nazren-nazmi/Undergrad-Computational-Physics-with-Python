```
#import libraries
import numpy as np
from matplotlib import pyplot as plt
```

```
def func(xls):
  if(xls < 0):
    return np.sin(xls*xls) + np.cos(xls*xls)
  elif(xls == 0):
    return 0
  else:
    return np.sin(xls*xls) - np.cos(xls*xls)

xls = np.linspace(-np.pi,np.pi,100)
yls = []
for i in range(len(xls)):
  yls.append(func(xls[i]))

plt.plot(xls,yls)
plt.plot(xls,yls,".")
plt.title('Plot of y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```
from mpl_toolkits.mplot3d import Axes3D
```

```
xls = []
yls = []
zls = []

for i in np.linspace(-np.pi,np.pi,100):
  for j in np.linspace(-np.pi,np.pi,100):
    xls.append(i)
    yls.append(j)
    zls.append((j**2)*np.sin(i) - (i**2)*np.cos(i))

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_trisurf(xls,yls,zls,cmap='coolwarm')
plt.title('Plot of z as a function of x and y')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
plt.show()

print('Maximum: %s' %max(zls))
print('Minimum: %s' %min(zls))
```
