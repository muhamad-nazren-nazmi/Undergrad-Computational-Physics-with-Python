```
#Name: Muhamad Nazren Nazmi Bin Muhamad Rusli
#Matric No: 1718151

# Import libraries
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import lagrange,CubicSpline
```

```
# Original data: x and f(x)
xls = [0.05, 0.20, 0.50, 0.750]
yls = [-4.813E-04, -2.720E-02, -3.125E-01, -7.383E-01]
```

```
# Scatter plot of the original data
plt.scatter(xls, yls)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Scatter plot of f(x) vs x')
```

```
#CubicSpline interpolations
cubsp_notaknot_func = CubicSpline(xls,yls,bc_type='not-a-knot')
cubsp_clamped_func = CubicSpline(xls,yls,bc_type='clamped')
cubsp_natural_func = CubicSpline(xls,yls,bc_type='natural')
```

```
#New x and f(x) data by interpolations
new_xls = np.linspace(0,1,11)

cubsp_notaknot_yls = cubsp_notaknot_func(new_xls)
cubsp_clamped_yls = cubsp_clamped_func(new_xls)
cubsp_natural_yls = cubsp_natural_func(new_xls)
```

```
#Plot of f(x) vs x by interpolations
plt.scatter(xls, yls)

plt.plot(new_xls, cubsp_notaknot_yls, label='CubicSpline(not-a-knot)')
plt.plot(new_xls, cubsp_clamped_yls, label='CubicSpline(clamped)')
plt.plot(new_xls, cubsp_natural_yls, label='CubicSpline(natural)')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) vs x by interpolations')
plt.legend()
```

```
#First derivative by gradient method

grad_notaknot_1d = np.gradient(cubsp_notaknot_yls,new_xls)
grad_clamped_1d = np.gradient(cubsp_clamped_yls,new_xls)
grad_natural_1d = np.gradient(cubsp_natural_yls,new_xls)
```

```
#Plot of dy/dx vs x by gradient method

plt.plot(new_xls, grad_notaknot_1d, label='CubicSpline(not-a-knot)')
plt.plot(new_xls, grad_clamped_1d, label='CubicSpline(clamped)')
plt.plot(new_xls, grad_natural_1d, label='CubicSpline(natural)')

plt.xlabel('x')
plt.ylabel('dy/dx')
plt.title('First derivative by gradient method')
plt.legend()
```

```
# Calculation of the mean absolute error by gradient method
true_xls = [0.1,0.4,0.7]
true_1d_yls = [-1.080E-01, -1.152, -1.764]
print(new_xls)

grad_notaknot_MAE =(abs(grad_notaknot_1d[1] - true_1d_yls[0]) + abs(grad_notaknot_1d[4] - true_1d_yls[1]) + abs(grad_notaknot_1d[7] - true_1d_yls[2]))/3.0
print('Mean absolute error by gradient method from cubic spline (not-a-knot) interpolation: %s' %grad_notaknot_MAE)

grad_clamped_MAE =(abs(grad_clamped_1d[1] - true_1d_yls[0]) + abs(grad_clamped_1d[4] - true_1d_yls[1]) + abs(grad_clamped_1d[7] - true_1d_yls[2]))/3.0
print('Mean absolute error by gradient method from cubic spline (clamped) interpolation: %s' %grad_clamped_MAE)

grad_natural_MAE =(abs(grad_natural_1d[1] - true_1d_yls[0]) + abs(grad_natural_1d[4] - true_1d_yls[1]) + abs(grad_natural_1d[7] - true_1d_yls[2]))/3.0
print('Mean absolute error by gradient method from cubic spline (natural) interpolation: %s' %grad_natural_MAE)
```

```
#First derivative by CubicSpline method
CS_notaknot_1d = cubsp_notaknot_func.derivative(nu=1)
CS_notaknot1d_yls = CS_notaknot_1d(new_xls)

CS_clamped_1d = cubsp_clamped_func.derivative(nu=1)
CS_clamped1d_yls = CS_clamped_1d(new_xls)

CS_natural_1d = cubsp_natural_func.derivative(nu=1)
CS_natural1d_yls = CS_natural_1d(new_xls)
```

```
#Plot of dy/dx by CubicSpline method
plt.plot(new_xls,CS_notaknot1d_yls,label='CubicSpline(not-a-knot)')
plt.plot(new_xls,CS_clamped1d_yls,label='CubicSpline(clamped)')
plt.plot(new_xls,CS_natural1d_yls,label='CubicSpline(natural)')

plt.xlabel('x')
plt.ylabel('dy/dx')
plt.title('First derivative by CubicSpline method')
plt.legend()
```

```
# Calculation of the mean absolute error by CubicSpline method
true_xls = [0.1,0.4,0.7]
true_1d_yls = [-1.080E-01, -1.152, -1.764]
print(new_xls)

CS_notaknot_MAE =(abs(CS_notaknot1d_yls[1] - true_1d_yls[0]) + abs(CS_notaknot1d_yls[4] - true_1d_yls[1]) + abs(CS_notaknot1d_yls[7] - true_1d_yls[2]))/3.0
print('Mean absolute error by CubicSpline method from cubic spline (not-a-knot) interpolation: %s' %CS_notaknot_MAE)

CS_clamped_MAE =(abs(CS_clamped1d_yls[1] - true_1d_yls[0]) + abs(CS_clamped1d_yls[4] - true_1d_yls[1]) + abs(CS_clamped1d_yls[7] - true_1d_yls[2]))/3.0
print('Mean absolute error by CubicSpline method from cubic spline (clamped) interpolation: %s' %CS_clamped_MAE)

CS_natural_MAE =(abs(CS_natural1d_yls[1] - true_1d_yls[0]) + abs(CS_natural1d_yls[4] - true_1d_yls[1]) + abs(CS_natural1d_yls[7] - true_1d_yls[2]))/3.0
print('Mean absolute error by CubicSpline method from cubic spline (natural) interpolation: %s' %CS_natural_MAE)
```
