```python
import numpy as np
from IPython.display import Audio
from scipy.io import wavfile
from matplotlib import pyplot as plt

from scipy import fftpack
```

```python
sampleaudiopath = '/content/drive/MyDrive/Colab Notebooks/Last Sem/samplesound.wav'
fs, sample = wavfile.read(sampleaudiopath)
tls = np.linspace(0,0.02,len(sample))
```

```python
Audio(sample,rate=fs)
```

```python
tls = np.linspace(0,0.02,len(sample))
plt.plot(tls, sample)
```

```python
magFls = fftpack.fft(sample)
freqls = fftpack.fftfreq(len(sample), 1/fs)
fmask = np.where(freqls > 0)
```

```python
plt.plot(freqls[fmask], abs(magFls[fmask]))
plt.xlim(0,1000) #domain showm is between 0 to 1000 only
```

```python
lowpass = 200
highpass = 800

filterfreqls = magFls * (abs(freqls) < highpass)
filterfreqls = filterfreqls * (abs(freqls) > lowpass)
```

```python
plt.plot(freqls[fmask], abs(filterfreqls[fmask]))
```

```python
filteredsound = fftpack.ifft(filterfreqls)
```

```python
plt.plot(tls, filteredsound)
```

```python
Audio(filteredsound,rate=fs)
```
