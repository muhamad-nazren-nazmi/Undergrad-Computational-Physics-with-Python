```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
sample = np.random.gamma(3, size = (1000))
sample2 = np.random.normal(15.3, 2, size=(1000))
allsample = np.append(sample, sample2)
```

```python
plt.hist(allsample, 30)
plt.show()
```

```python
from scipy.stats.kde import gaussian_kde

kdemdl = gaussian_kde(allsample, bw_method=0.3)
xls = np.linspace(0, 17, 100)
kdescr = kdemdl(xls)
plt.plot(xls, kdescr)
plt.show()
```

```python
newsample = kdemdl.resample(2000)[0]
```

```python
plt.hist(newsample, 30)
plt.show()
```
